#### Initializing cowin

```python
from cowin_api import CoWinAPI

cowin = CoWinAPI()
```

#### Getting all states

```python
states = cowin.get_states()
print(states)
```

#### Getting available districts for Delhi

```python
state_id = '9'
districts = cowin.get_districts(state_id)
print(districts)
```

#### Getting all available centers in 'North West Delhi'



```python
district_id = '143'

# here date is today's date by default and no min_age_limit specified
available_centers = cowin.get_availability_by_district(district_id)
print(available_centers)
```

#### Getting all available centers in 'North West Delhi' for Age 18+

```python
district_id = '143'

# here date is today's date by default
available_centers = cowin.get_availability_by_district(district_id, min_age_limt=18)
print(available_centers)
```

#### Getting all available centers by pincode 110085

```python
# here date is today's date by default and no min_age_limit specified
available_centers = cowin.get_availability_by_pincode(pin_code='110085')
print(available_centers)
```
